```python
import re
import numpy as np

def runtime_earn(runtime):
  if pd.isna(runtime):
    return np.nan
  else:
    return re.sub('\D', '', runtime)

dataset['runtime'] = dataset['runtime'].apply(runtime_earn)
dataset['runtime']
```

```python
dataset['runtime_clear'] = dataset['runtime'].astype(int)
dataset['runtime_clear']
```
